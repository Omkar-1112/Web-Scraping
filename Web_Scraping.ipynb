```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
#location
#Price per area
#rental Yeild
def scrap_housingcom(filename):
    headers={"User-Agent": "Mozilla/5.0 (X11; CrOS x86_64 12871.102.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.141 Safari/537.36"}
    wp=requests.get('https://housing.com/rent/search-Psrbcrw7p4zhuczr',headers=headers).text
    
    html=BeautifulSoup(wp,'lxml')
    
    Rental_Yeild=[]
    Location=[]
    Price=[]
    for i in html.find_all('div',class_='T_091c165f _sq1l2s _vv1q9c _ks15vq T_3d3547ab _7s5wglyw _5vy24jg8 _blas1v10 new-title'):
        x=i.text.strip()
        temp=x.split(' ')[0]+' '+x.split(' ')[1]
        Rental_Yeild.append(temp)
    
    for i in html.find_all('div',class_='T_091c165f _sq1l2s _vv1q9c _ks15vq T_3d3547ab _7s5wglyw _5vy24jg8 _blas1v10 new-title'):
        x=i.text.strip()
        text=x.split()
        index=text.index('in')
        try:
            temp=x.split(' ')[index+1]+' '+x.split(' ')[index+2]
            Location.append(temp)
        except:
            temp=x.split(' ')[index+1]
            Location.append(temp)
    
    for i in html.find_all('div',class_="_csbfng _c8exct _g3qslr _fr1nms _1q73f6fq _4okugktf _1qkcc65e _j3gpidpf _4nrxyh40 T_a6707275"):
        Price.append(i.text.strip())
    
    df_SrapData = pd.DataFrame({
    'Location': Location,
    'Rental_Yeild': Rental_Yeild,
    'Price': Price
    })
    df_SrapData.to_csv(filename)
    return df_SrapData
```

```python
scrap_housingcom('demo.csv')
```
